```python
import requests
from bs4 import BeautifulSoup
```

```python
# 1. 웹 페이지 내용 가져오기
url = "https://newsstand.naver.com/?list&pcode=030"
response = requests.get(url)

# 2. HTML 파싱하기
soup = BeautifulSoup(response.text, 'html.parser')

# 3. 데이터 추출 (예: 모든 h1 태그 가져오기)
title = soup.find('h1').text
print(title)
```

```python
url = 'https://kin.naver.com/search/list.nhn?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC'

response = requests.get(url)

# 응답코드가 200일때
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    # 하나의 html 요소를 선택하여 텍스트 추출 / css 선택자 사용 / copy selector 기능이용
    title = soup.select_one('#s_content > div.section > ul')
    print(title)
else : 
    print(response.status_code)
```

```python
url = 'https://kin.naver.com/search/list.nhn?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    # ul.basic1 을 선택하여 그 안의 li > dl > dt > a 요소들을 모두 선택하여 텍스트 추출
    ul = soup.select_one('ul.basic1')
    # 리스트형태로 반환
    titles = ul.select('li > dl > dt > a')
    for title in titles:
        print(title.get_text())
else : 
    print(response.status_code)
```
