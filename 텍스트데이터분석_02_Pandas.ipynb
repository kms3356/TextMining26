### 시리즈 객체

생성: 리스트, 딕셔너리 등 시퀀스형 데이터로부터 생성할 수 있습니다.


일반적 방법: 보통 데이터프레임을 먼저 만들고 특정 열을 추출하여 시리즈를 얻는 방식이 많이 쓰입니다.


주요 속성: 


s.dtype: 데이터 타입 반환 


s.shape: 데이터 구조를 튜플로 반환 


s.size: 요소의 총 개수 반환 


s.name: 시리즈의 이름 반환 


s.index: 인덱스 반환 


s.values: 값을 넘파이 배열로 반환

```python
# 기본 문법
# series = pd.Series(sequence, dtype, index, name)

# 리스트로부터 생성
numbers = [1, 2, 3, 4, 5]
list_sr = pd.Series(numbers)

# 속성 확인
print(list_sr.dtype, list_sr.shape, list_sr.size)
print(list_sr.index)
print(list_sr.values)
print(list_sr)
```

```python
# 딕셔너리로부터 생성
dict_data = {"a":1, "b":2, "c":3, "d":4, "e":5}
dict_sr = pd.Series(dict_data)
print(dict_sr.dtype, dict_sr.shape, dict_sr.size)
print(dict_sr.index)
print(dict_sr.values)
print(dict_sr)
```

```python
# dtype 및 속성 지정
sr1 = pd.Series(numbers, dtype=float) 
import numpy as np
sr2 = pd.Series(numbers, dtype=np.float32)
print(sr2)
# 속성 수정 및 추가
sr1.index = ['a', 'b', 'c', 'd', 'e'] 
sr1.name = 'MySeries'
sr1
# 컬럼명 및 타입 변경 문법
# new_df = org_df.rename(columns={기존이름: 새이름}) 
# new_df = org_df.astype(columns={컬럼명: 타입}) 
```

```python
chars = 'a b c d e'.split()
sr3 = pd.Series(numbers, index=chars, name='MySeries')
sr3
```

```python
sr3.index, sr3.values
```

```python
# 딕셔너리로 만들기
s = pd.Series({'a':1,'b':2,'c':3,'d':4, 'e':5})
print(s.dtype, s.shape, s.size, s.name)
print(s.values)
print(s)
```

```python
numbers = [1,2,3,4,5]
pd1 = pd.DataFrame(numbers)
pd1
```

```python
# 2차원 리스트로부터 생성
score = [['정약용', 70, 80, 100],
         ['이순신', 100, 80, 70],
         ['이율곡', 90, 95, 98]]
score_df = pd.DataFrame(score)
score_df
```

```python
score_df.columns = ['이름', '영어', '수학', '국어']
score_df.index = ['a', 'b', 'c']
score_df
```

```python
print(score_df.shape, score_df.index, score_df.columns, score_df.dtypes, score_df.values, sep='\n')
```

```python
# 딕셔너리로부터 생성
score_dict = {'이름': ['정약용', '이순신', '이율곡', '홍길동'],
              '국어': [70, 100, 90, 85],
              '영어': [80, 80, 95, 100],
              '수학': [100, 70, 90, 90]}
score_df2 = pd.DataFrame(score_dict)
# 수정 예시 - 객체 반환하는 함수들임. 원본적용 x, 하려면 inplace=True 옵션 넣으면됨
score_df2.rename(columns={'이름':'name', '국어':'kerean'})
score_df2.astype({'국어': float}) 
score_df2['국어'].dtype
```

```python
# 열 추가 및 연산
score_df2['사회'] = [85, 88, 92, 96]
# score_df2['총점'] = score_df2.국어 + score_df2.영어 + score_df2.수학 + score_df2.사회 이렇게 해도됨 이미 있는 컬럼만.
score_df2['총점'] = score_df2['국어'] + score_df2['영어'] + score_df2['수학'] + score_df2['사회']
score_df2['평균'] = score_df2.총점 / 4
score_df2
```

```python
# 수식 기반 열 추가 예시
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]
df = pd.DataFrame()
df['x'] = x 
df['y'] = y 
df['x*y'] = df['x'] * df['y']
df
```

```python
# 삭제 및 초기화 (inplace=True)
score_df.drop(['이름'], axis=1) # 열 삭제 - 컬럼명으로 삭제
score_df.drop(['a', 'b'], axis=0) # 행 삭제 - 인덱스로 삭제 디폴트는 행
score_df.reset_index(drop=True) # 인덱스 초기화 및 기존 인덱스 삭제 
```

```python
# 열(column) 접근 - select
score_df2['이름']
score_df2.이름
print(score_df2[['이름', '총점', '평균']]) #boolean 인덱싱 비슷한 느낌
# 행 슬라이싱. 행하나로 접근은 안됨
df[1:3]
```

```python
# Boolean Indexing(행 추출) - where 조건걸기
score_df2[score_df2['수학'] > 80]
score_df2[(score_df2['수학'] > 80) & (score_df2['영어'] < 90)]
score_df2[score_df2['이름'].isin(['정약용', '이순신'])]
```

```python
# df.loc, df.iloc
print(score_df2.loc[0])
print(score_df2.loc[0,'국어'])
print(score_df2.loc[0, ['이름', '총점']])
print(score_df2.loc[0:2, '국어':'수학']) # 끝점포함

print(score_df2.iloc[0,0])
print(score_df2.iloc[0:1,0:3]) # 끝점제외
```

외부 파일 로딩다양한 형식의 파일을 판다스 객체로 변환하여 분석을 시작합니다.

CSV 로드: pd.read_csv('파일경로', sep=',', encoding=...) 형식을 사용합니다.

JSON 로드: pd.read_json('파일경로', encoding=...) 형식을 사용합니다.

데이터 확인을 위한 주요 메서드데이터의 전체적인 구조와 누락된 값(결측치) 등을 빠르게 파악할 때 사용합니다.

앞/뒤 데이터 확인: head(n)는 상위 5개(기본값), tail(n)은 하위 5개의 행을 출력합니다.

크기 및 개수: shape는 (행, 열)의 크기를 튜플로 반환하며, len()은 전체 데이터의 개수(행의 수)를 알려줍니다.

정보 요약: info()를 통해 각 열의 데이터 개수와 결측치(null, NaN) 존재 여부, 데이터 타입을 한눈에 확인합니다.

기술 통계: describe()는 숫자형(float, int) 열에 대해서만 평균, 표준편차, 최솟값, 최댓값 등의 통계치를 계산해 줍니다.

값별 빈도: value_counts()는 특정 열의 데이터 값들이 각각 몇 개씩 들어있는지 집계합니다.

```python
import pandas as pd

# 1. 파일 불러오기 [cite: 440]
sample_df = pd.read_csv('./data/sample_data.csv')

# 2. 상단 및 하단 데이터 확인 (각 3개씩)
print(sample_df.head(3))
print(sample_df.tail(3))

# 3. 데이터프레임 구조 확인
print("구조(행, 열):", sample_df.shape)
print("전체 데이터 개수:", len(sample_df))

# 4. 컬럼별 정보 및 결측치 확인
sample_df.info()
```

```python
# 5. 수치형 데이터 통계 요약
print(sample_df.describe())

# 6. 특정 컬럼의 값 빈도수 확인 (예: 'City' 컬럼) 
sample_df['City'].value_counts()
```

```python
# 결측치
# 컬럼별 결측치 갯수
sample_df.isnull().sum() 

# 인덱스 별 결측치 갯수
sample_df.isnull().sum(axis=1) 

# 인덱스 별 결측치 True or False
sample_df.isnull().any(axis=1) 

# boolean indexing
sample_df[sample_df.isnull().any(axis=1)] 

# 결측지 있는 행 삭제 (inplace=True) 하면 원본 바뀌지만 리턴값 없기 때문에 변수지정 x
sample_df.dropna() 

# 지정한 컬럼에서 결측치 있는 행 삭제
sample_df.dropna(subset=['City']) 
```

```python
# 중복치
# 중복 행 True or False
sample_df.duplicated() # default : 중복 행 중에 첫번째 남김

# 중복 행 추출
sample_df[sample_df.duplicated(keep='first')] # keep = 'first' : 첫번째 행 남김 'last' : 마지막 행 남김 False : 둘다 추출

# 중복 행 삭제
sample_df.drop_duplicates() # keep, inplace
```
