### 서울시 거주/비거주 고객 비교 분석

```python
seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] == '서울특별시']
be_seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] != '서울특별시']
con_seoul = pd.DataFrame(index=["서울시 거주", "서울시 비거주"], columns=["거주/비거주"])
con_seoul["고객수"] = [f"{len(seoul):,}", f"{len(be_seoul):,}"]
con_seoul[f"소비액"] = [f"{seoul['AMT'].sum():,}", f"{be_seoul['AMT'].sum():,}"]
con_seoul["이용건수"] = [f"{seoul['CNT'].sum():,}", f"{be_seoul['CNT'].sum():,}"]
con_seoul
```

### 편의점 소비 정보 분석

```python
ganpeon = bc_df[(bc_df['CTY_RGN_NM'] == '강남구') & (bc_df['TP_BUZ_NO'] == 4010)]
data = [[f"{bc_df[bc_df['TP_BUZ_NO'] == 4010]['AMT'].sum():,}", f"{bc_df[bc_df['CTY_RGN_NM'] == '강남구']['AMT'].sum():,}", 
         f"{ganpeon['AMT'].sum():,}"]]
peon_sobi = pd.DataFrame(data,index=["소비액"], columns=["편의점", "강남구", "강남구 편의점"])
peon_sobi
```

```python
data = [[f"{ganpeon['AMT'].sum():,}", f"{ganpeon[ganpeon['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum():,}", f"{ganpeon[ganpeon['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum():,}"]]
ganpeon_sobi = pd.DataFrame(data, index=["소비액"], columns=["강남구 편의점 총액", "강남구 거주자", "강남구 비거주자"])
ganpeon_sobi
```
