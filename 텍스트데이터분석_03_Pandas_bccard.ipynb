### 데이터 로딩 및 확인

```python
import pandas as pd
bc_df = pd.read_csv('./data/bccard_201906.csv')
# 컬럼명 확인
print(bc_df.columns)
```

### 앞부분, 뒷부분, 데이터 수

```python
print(bc_df.head(3))
print(bc_df.tail(3))
print(f"데이터 수 : {len(bc_df)}")
```

### 서울시 거주/비거주 고객 비교 분석

```python
seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] == '서울특별시']
be_seoul = bc_df[bc_df['CSTMR_MEGA_CTY_NM'] != '서울특별시']
print(f"서울시 거주 고객수 : {len(seoul):,} 명\n서울시 비거주 고객수 : {len(be_seoul):,} 명")
print(f"서울시 거주 총 소비액 : {seoul['AMT'].sum():,} 원\n서울시 비거주 총 소비액 : {be_seoul['AMT'].sum():,} 원")
print(f"서울시 거주 총 이용건수 : {seoul['CNT'].sum():,} 건\n서울시 비거주 총 이용건수 : {be_seoul['CNT'].sum():,} 건")
```

### 편의점 소비 정보 분석

```python
print(f"편의점 소비액 : {bc_df[bc_df['TP_BUZ_NO'] == 4010]['AMT'].sum():,} 원")
print(f"강남구 소비액 : {bc_df[bc_df['CTY_RGN_NM'] == '강남구']['AMT'].sum():,} 원")
ganpeon = bc_df[(bc_df['CTY_RGN_NM'] == '강남구') & (bc_df['TP_BUZ_NO'] == 4010)]
print(f"강남구 편의점 소비액 : {ganpeon['AMT'].sum():,} 원")
print(f"강남구 편의점 소비액 중 강남구 거주자 : {ganpeon[ganpeon['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum():,} 원")
print(f"강남구 편의점 소비액 중 강남구 비거주자 : {ganpeon[ganpeon['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum():,} 원")
```
